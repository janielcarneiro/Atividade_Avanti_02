```python
#Esse codigo vai exibir varios detalhes da imagem
import cv2
from matplotlib import pyplot as plt

#carregar imagem
caminho_imagem = '/content/drive/MyDrive/Atividade_Atlantico_2/eu.41.15.jpeg'
image = cv2.imread(caminho_imagem)

# Converter imagem
imagem = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#Normalizar a imagem ficar intervalo (0 e 1) os tons de cinza
imagem_normalizada = imagem / 255.0

#Fazer a exebição da imagem
plt.imshow(imagem_normalizada)
plt.axis('off') #desativa os eixos
plt.show()

#Logica para mostrar as propriedades
if imagem_normalizada is not None:
  #obtem dimensões da imagem
  height, width, channel = imagem_normalizada.shape

  #Lista todas as propriedades da imgem
  propriedades = [
      ('Altura', height),
      ('Largura', width),
      ('Canais de cor', imagem_normalizada.shape[1]),
      ('Tipo de dado', imagem_normalizada.dtype),
      ('Tom de Cinza Máximo', imagem_normalizada.max()),
      ('Tom de Cinza Médio', imagem_normalizada.mean()),
      ('Tom de Cinza Minimo', imagem_normalizada.min()),
  ]

  #Exibir todas as Propriedades
  for prop_name, prop_value in propriedades:
    print(f"{prop_name}: {prop_value}")
else:
  print("Erro ao carregar a imagem")
```
