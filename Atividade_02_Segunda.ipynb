```python
#FILTRO DA MÃDIA
#Carregando filtro da MÃ©dia
import cv2
from matplotlib import pyplot as plt

#Carregar a imagem
caminho_imagem = '/content/drive/MyDrive/Atividade_Atlantico_2/Cidade-da-gente.jpg'
imagem = cv2.imread(caminho_imagem)

#APLICAR O FILTRO DE MÃDIA

#tamanho da janela da media
blurred_image = cv2.blur(imagem, (5,5))

#plotar a imagem Media suavizada
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
plt.title('Colorida')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Imagem Suavizada(Filtro MÃ©dia)')
```

```python
#FILTRO DA GAUSSIANO

#Carregando filtro da MÃ©dia
import cv2
from matplotlib import pyplot as plt

#Carregar a imagem
caminho_imagem = '/content/drive/MyDrive/Atividade_Atlantico_2/Cidade-da-gente.jpg'
imagem = cv2.imread(caminho_imagem)

#APLICAR FILTRO GAUSSIANO

#tamanho da janela da media
blurred_image = cv2.GaussianBlur(imagem, (7,7), 0)

#plotar a imagem Media suavizada
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
plt.title('Colorida')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Imagem Suavizada(Filtro Gaussiano)')
```

```python
#Carregando filtro da MÃ©diana
import cv2
from matplotlib import pyplot as plt

#Carregar a imagem
caminho_imagem = '/content/drive/MyDrive/Atividade_Atlantico_2/Cidade-da-gente.jpg'
imagem = cv2.imread(caminho_imagem)

#APLICAR FILTRO MÃDIANA

#tamanho da janela da media
blurred_image = cv2.medianBlur(imagem, 5)

#plotar a imagem Media suavizada
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
plt.title('Colorida')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Imagem Suavizada(Filtro MÃ©diana)')
```
